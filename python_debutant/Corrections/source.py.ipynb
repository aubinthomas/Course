```python
###################################
# Volume du cône
# ex 1
###################################

# définition des variables
rayon = 5.2
hauteur = 14.3
pi = 3.14

# formule du volume avec approximation de pi
volume1 = (1/3)* pi * rayon**2 * hauteur

# formule du volume avec la variable pi de la lib math
import math
volume2 = (1/3)* math.pi * rayon**2 * hauteur

# la difference observee
dV = abs(volume2-volume1)

print("avec approximation, volume = ", volume1)
print("avec math.pi, volume = ", volume2)
print("soit une différence de ",dV)
```

```python
###################################
# Numpy arrays
# ex 5
###################################
import pandas as pd
data = pd.read_csv("https://github.com/dbendet/coursera_machine_learning/raw/master/kc_house_data.csv")
print(data)
print()
data = data.to_numpy() # on transforme l'objet dataframe en array numpy

#### combien de lignes et colonnes
dimensions = data.shape
print(type(dimensions))
nbl = dimensions[0] # nombre de lignes
nbc = dimensions[1] # nombre de colonnes

print("lignes = ",nbl)
print("colonnes = ",nbc)
print()

#### Calcul du Z-score du prix (3e colonne=>indice 2)
priceMean = data[:,2].mean()  #valeur mean des prix
priceStd  = data[:,2].std()   #stdandard deviation des prix
priceZScore = (data[:,2]-priceMean)/priceStd
print("Z-score :")
print(priceZScore)
print()


#### outlier
# on fait un sort des valeurs pour identifier les valeurs correspondant aux 2% des valeurs extrêmes
priceCol = data[:,2].copy()
priceColSorted = priceCol.copy()
priceColSorted.sort()
val5percentMin = priceColSorted[int(nbl*0.02)]
val5percentMax = priceColSorted[int(nbl*0.98)]
#toutes les valeurs qui ne sont pas des outliers sont comprises entre val5percentMin et val5percentMax
np.bincount(np.greater(priceCol,val5percentMin) &  np.less(priceCol,val5percentMax))  #false est vu comme 0, true comme 1

```

```python
###################################
# boucles + dict + numpy
# ex 9
###################################

import pandas as pd
data = pd.read_csv("https://github.com/dbendet/coursera_machine_learning/raw/master/kc_house_data.csv")
print(data)
data = data.to_numpy()

def Score(price, surface):
    score = price/surface
    return score


# score max
numberOfLines = data.shape[0]
line = 0
result = {}

while(line<numberOfLines):
    numberOfRooms = data[line,3] + data[line,4]
    score_current = Score(data[line,2],data[line,5])
    if(numberOfRooms in result):
        if(result[numberOfRooms]<score_current):
            result[numberOfRooms]=score_current
    else:
        result[numberOfRooms]=score_current
    line = line+1

print(result)



```
