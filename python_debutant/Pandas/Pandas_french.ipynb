## loc

### <b> Trouver toutes les lignes selon des conditions sur une colonne </b> 

```python
df.loc[df.age>20]
```

```python
df.loc[(df.age>20) | (df.salary>500)]
```

### <b>Trouver un ensemble de lignes avec loc</b>

```python
df.loc[1:3]
```

```python
df = df.sort_values(by='age')
df 
```

```python
df.loc[1]
```

### <b>Sélectionner des colonnes avec une condition</b>

```python
df.loc[df.age>12,['city', 'gender']]
```

### <b>update des valeurs selon condition</b>

```python
#update 1 column
df.loc[(df.age> 12)&(df.gender=='F'),['section']] = "NM"
df
```

```python
#update 2 columns
df.loc[(df.age> 12)&(df.gender=='F'),['section','salary']] = ["NM",14250]
df
```
